```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
historical_data = pd.read_csv('historical_data.csv')
historical_data.head()
```

```python
fear_greed_index = pd.read_csv('fear_greed_index.csv')
fear_greed_index.head()
```

```python
# Step 1: Preprocessing

fear_greed_index['date'] = pd.to_datetime(fear_greed_index['date'])

# Convert timestamp in historical_data to datetime
historical_data['Timestamp IST'] = pd.to_datetime(historical_data['Timestamp IST'], format='%d-%m-%Y %H:%M')

# Extract date from historical_data timestamp
historical_data['date'] = historical_data['Timestamp IST'].dt.date
historical_data['date'] = pd.to_datetime(historical_data['date'])  # Ensure datetime type

# Step 2: Aggregate historical_data performance by date
historical_daily = historical_data.groupby('date').agg({
    'Closed PnL': 'sum',
    'Execution Price': 'mean',
    'Size USD': 'sum',
    'Account': 'nunique',  # Number of unique traders
    'Side': 'count'        # Number of trades
}).reset_index().rename(columns={
    'Closed PnL': 'Total PnL',
    'Execution Price': 'Avg Execution Price',
    'Size USD': 'Total Trade Volume',
    'Account': 'Unique Traders',
    'Side': 'Trade Count'
})

# Step 3: Merge with fear_greed_index data
merged_df = pd.merge(historical_daily, fear_greed_index, on='date', how='inner')

# Display merged result
merged_df.head()
```

```python
q = merged_df['Avg Execution Price'].quantile(0.99)
data_1 = merged_df[merged_df['Avg Execution Price']<q]
data_1.describe(include='all')
```

```python
sns.displot(data_1['Avg Execution Price'])
plt.savefig('avgexecution.jpg')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Plot 1: Total PnL vs Sentiment Score over time
plt.figure(figsize=(14, 6))
ax1 = sns.lineplot(data=merged_df, x="date", y="Total PnL", label="Total PnL", color="blue")
ax2 = ax1.twinx()
sns.lineplot(data=merged_df, x="date", y="value", label="Sentiment Index", color="orange", ax=ax2)
ax1.set_title("Trader PnL vs Market Sentiment Over Time")
ax1.set_ylabel("Total PnL")
ax2.set_ylabel("Sentiment Index")
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
plt.savefig('totalpnl vs fear_greed_inedex')
```

 Correlation heatmap

```python
correlation_matrix = merged_df[['Total PnL', 'Avg Execution Price', 'Total Trade Volume', 'Unique Traders', 'Trade Count', 'value']].corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Trader Metrics and Sentiment Index")
plt.show()
plt.savefig('Correlation between trader')
```

```python
# Scatter plot of Total PnL by Sentiment classification

plt.scatter(x=merged_df['classification'], y=merged_df['Total PnL'])
plt.title("Distribution of Trader PnL by Market Sentiment Classification")
plt.xlabel("Sentiment Classification")
plt.ylabel("Total PnL")
plt.show()
plt.savefig('Distribution of Trader PnL')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Our algorithm find optimal coefficients of a linear regression model

x = merged_df[['value', 'Avg Execution Price', 'Total Trade Volume', 'Unique Traders', 'Trade Count', 'classification']]
y = merged_df['Total PnL']
```

```python
x.shape
```

```python
y.shape
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.score(x,y)
```

```python
# Coefficient

reg.coef_
```

```python
# Intercept

reg.intercept_
```

```python
# prediction

reg.predict(x)
```

```python
# Scatter plot of actual vs predicted

plt.figure(figsize=(8,5))
plt.scatter(y, reg.predict(x), alpha=0.7, color='blue')
plt.xlabel("Actual Total PnL")
plt.ylabel("predicted Total PnL")
plt.title("Actual vs Predicted Total PnL")
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.grid(True)
plt.show()
plt.savefig('Actual vs Predicted')
```
